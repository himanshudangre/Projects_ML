```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
df = pd.read_excel(open('KPMG.xlsx', 'rb'),sheet_name='CustomerDemographic')  
df
```

```python
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
df
```

```python
df.isnull().sum()
```

```python
df['job_title']=df['job_title'].fillna(method='ffill')
df
```

```python
df['last_name']=df['last_name'].fillna(method='ffill')
```

```python
df['DOB']=df['DOB'].fillna(method='ffill')
```

```python
df['job_industry_category'].value_counts()
```

```python
df['job_industry_category']=df['job_industry_category'].fillna(method='ffill')
```

```python
del df['default']
```

```python
df['tenure'].hist()
```

```python
df['tenure']=df['tenure'].fillna(df['tenure'].mean())
```
