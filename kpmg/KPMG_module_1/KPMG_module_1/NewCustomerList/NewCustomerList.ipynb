```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pandas import ExcelWriter
from pandas import ExcelFile

df = pd.read_excel(open('KPMG.xlsx', 'rb'),sheet_name='NewCustomerList')  
df
```

```python
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
df
```

```python
df.isnull().sum()
```

```python
df['last_name']=df['last_name'].fillna(method='ffill')
```

```python
df['DOB']=df['DOB'].fillna(method='ffill')
```

```python
df['job_title']=df['job_title'].fillna(method='bfill')
```

```python
df['job_industry_category']=df['job_industry_category'].fillna(method='bfill')
```

```python
df.head()
```

```python
temp=pd.DataFrame(df["Rank"])
temp["Value"]=df["Value"]
temp
```

```python
df.drop(df.iloc[:, 16:], inplace = True, axis = 1)
df["Rank"]=temp["Rank"]
df["Value"]=temp["Value"]
df
```

```python
df.to_csv("cleanedNewCustomerList.csv")
```
