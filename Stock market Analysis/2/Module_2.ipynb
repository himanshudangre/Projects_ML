### Task 2.9

```python
sma=pd.DataFrame(dfidfc["Date"])
sma["Average Price"]=dfidfc["Average Price"]
sma["Close Price"]=dfidfc["Close Price"]
sma["Average Price"]=dfidfc["Close Price"]
```

```python
sma["Roll21"]=sma["Close Price"].rolling(window=21,min_periods=1, center=False).mean()
sma["Roll34"]=sma["Close Price"].rolling(window=34,min_periods=1, center=False).mean()
```

```python
sma["val"]=0.0
sma['ind'] = np.where(sma['Roll21'] > sma['Roll34'], 1.0,0.0)
```

```python
sma["val"]= sma["ind"].diff()
```

```python
sma["val"].value_counts()
```

```python
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(111, ylabel='Price')
sma['Close Price'].plot(ax=ax1, color='black', lw=2.)
sma[['Roll21', 'Roll34']].plot(ax=ax1, lw=2.)
ax1.plot(sma.loc[sma.val == 1.0].index, sma.Roll21[sma.val == 1.0], '^' , markersize=20,color='g')
ax1.plot(sma.loc[sma.val ==-1.0].index, sma.Roll21[sma.val == -1.0], 'v' , markersize=20,color='r')
plt.show()
```

BOLU=MA(TP,n)+m∗σ[TP,n]
BOLD=MA(TP,n)−m∗σ[TP,n]
The closer the prices move to the upper band, the more overbought the market, and the closer the prices
move to the lower band, the more oversold the market.
A squeeze represents increase in volatitlity in future, potential sign for increased trade.
As band widens it represents decrease in volality, grater chance of exiting the trade.

```python
dra=dfpvr["Close Price"].rolling(14).mean()
up=dra+2*dfpvr["Close Price"].rolling(14).std()
down=dra-2*dfpvr["Close Price"].rolling(14).std()
```

```python
plt.figure(figsize=(20,10))
plt.plot(dfpvr["Date"],dra,color="black",linewidth=2)
plt.plot(dfpvr["Date"],up)
plt.plot(dfpvr["Date"],down)
plt.plot(dfpvr["Date"],dfpvr["Close Price"],color="red")
plt.xlabel("Date")
plt.ylabel("sma and Boliinger plots")
plt.title("Bollinger Plots")
plt.show()
```

```python
dfpvr["dra"]=dra
dfpvr["up"]=up
dfpvr["down"]=down
```
